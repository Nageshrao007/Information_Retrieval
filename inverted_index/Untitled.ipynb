```python
import glob
import os
import re
from nltk.stem import PorterStemmer as ps 
import nltk

numbers = re.compile(r'(\d+)')
def numbersorting(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

def remove_punc(data):
    for p in "!.,:@#$%^&?<>*()[}{]-=;/\"\\\t\n":
        if p in '\n;?:!.,.':
            data = data.replace(p,' ')
        else: data = data.replace(p,'')
    return data.lower()

list_docs = []
for file in sorted(glob.glob('file_33.txt'),key = numbersorting):
    with open(file,'r') as f:
        if file not in list_docs:
            list_docs.append(os.path.basename(file))

        
```

```python
pos_index = []
newlist = []
for i in range(len(list_docs)):
    with open(list_docs[i]) as f:
        doc = [a for a in remove_punc(f.read()).split(' ') if a != ""]
        pos_index.append(tf(doc))
```

```python
def tf(corpus):
    tfs = []
    for document in corpus:
        dic={}
        for word in document.split():
            if word in dic:
                dic[word]+=1
            else:
                dic[word]=1
        for word,freq in dic.items():
            print(word,freq)
            dic[word]=freq/len(document.split())
        tfs.append(dic)
    return tfs
```
