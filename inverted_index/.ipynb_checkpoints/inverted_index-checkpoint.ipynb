```python
import glob
from nltk.stem import PorterStemmer
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
import re
```

```python
#to extract the file number from file name
def numbers_from_filename(filename):
    return re.search(r'\d+', filename).group(0)

#to remove puncutations 
def remove_punc(string):
    punc = '''!()-[]{};:'"\, <>.``/?@#$%^&*_~'''
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, "") 
    return string
```

```python
wordlist = {}
for file in glob.glob('file*.txt'):
    with open(file,'r') as f:
        text = f.read()
        #print(file)
        file = numbers_from_filename(file)
        toks = word_tokenize(text)
        stemmer = nltk.stem.PorterStemmer()
        stemmed_words = [stemmer.stem(word) for word in toks]
        stop_words = list(set(stopwords.words('english')))
        newlist = []
        for r in stemmed_words:
            if r not in stop_words:
                newlist.append(r)
        for i in newlist:
            if i not in wordlist.keys():
                wordlist[i] = {}
                wordlist[i] = file
            else:
                wordlist[i] = wordlist[i] + "," + file
```

```python
import json
with open("invertedindex.json", "w") as outfile:
    json.dump(wordlist, outfile)
```
